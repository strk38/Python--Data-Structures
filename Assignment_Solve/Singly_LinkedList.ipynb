```python
class Node:
  def __init__(self,elem,next=None):
    self.elem=elem
    self.next=next
```

```python
class LinkedList:
  def __init__(self):
    self.head=None

  def push(self,data):
    new_node=Node(data)
    new_node.next=self.head
    self.head=new_node

  def display(self):
    current=self.head
    while current:
      print(current.elem, end='->')
      current=current.next
    print('None')

  def moveFromLast(self):
    if self.head is None or self.head.next is None:
      return

    head=self.head
    prev=head.next
    last=head.next

    while last.next is not None:
      prev=last
      last=last.next
    prev.next=None
    last.next=head
    self.head=last

linked_list = LinkedList()

# Add some nodes to the linked list
linked_list.push(3)
linked_list.push(7)
linked_list.push(9)
linked_list.push(2)
linked_list.push(5)

print("Original Linked List:")
linked_list.display()

# Move the last node to the front
linked_list.moveFromLast()

print("\nLinked List after moving the last node to the front:")
linked_list.display()
```

```python
class Node:
  def __init__(self,elem,next=None):
    self.elem=elem
    self.next=next

class LinkedList:
  def __init__(self):
    self.head=None

  def push(self,data):
    node=Node(data)
    if not self.head:
      self.head=node
    else:
      tail=self.head
      while tail.next:
        tail=tail.next
      tail.next=node

  def display(self):
    current=self.head
    while current:
      print(current.elem,end='->')
      current=current.next
    print('None')

def FetchEven(list1):
  current=list1.head
  even=None
  even_tail=None
  odd=None
  odd_tail=None

  while current is not None:
    new_node=Node(current.elem)
    if current.elem%2==0 :
      if even is None:
        even=new_node
        even_tail=new_node
      else:
        even_tail.next=new_node
        even_tail=new_node

    else:
      if odd is None:
        odd=new_node
        odd_tail=new_node
      else:
        odd_tail.next=new_node
        odd_tail=new_node
    current=current.next

  if even_tail:
    even_tail.next=odd

  lst=LinkedList()
  lst.head=even
  return lst


linked_list1 = LinkedList()

# Add some nodes to the linked list
linked_list1.push(2)
linked_list1.push(18)
linked_list1.push(24)
linked_list1.push(3)
linked_list1.push(5)
linked_list1.push(7)
linked_list1.push(9)
linked_list1.push(6)
linked_list1.push(16)
print("Original Linked List:")
linked_list1.display()

lst=FetchEven(linked_list1)
lst.display()
```
